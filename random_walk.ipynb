```python
# Python code for 1-D random walk.
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
def random_walk_1d(steps):
   """Generate a 1D random walk."""
   walk = np.zeros(steps)
   for i in range(1, steps):
      step = np.random.choice([-1, 1])
      walk[i] = walk[i - 1] + step
   return walk

# Number of steps
steps = 1000
walk = random_walk_1d(steps)

# Plot the random walk
plt.figure(figsize=(10, 6))
plt.plot(walk, label='1D Random Walk')
plt.xlabel('Steps')
plt.ylabel('Position')
plt.title('1D Random Walk')
plt.legend()
plt.show()
```

```python
def random_walk_2d(steps):
    positions = np.zeros((steps,2))
    for i in range(1,steps):
        step = np.random.choice(['up', 'down', 'left', 'right'])
        if step == 'up':
            positions[i] = positions[i-1] + [0,1]
        elif step == 'down':
            positions[i] = positions [i-1] + [0,-1]
        elif step == 'left':
            positions[i] = positions [i-1] + [-1,0]  
        elif step == 'right':
            positions [i] = positions [i-1] + [1,0]
    return positions

steps = 1000
coordinates = random_walk_2d (steps)
print(coordinates)
plt.figure(figsize=(10,10))
plt.plot(coordinates[:,0], coordinates[:,1], label = '2D random Walk')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('2D Random Walk')
plt.legend()
plt.grid(True)
plt.show()

            
```
